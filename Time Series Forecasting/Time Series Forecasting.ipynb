```python
# Step 1: Prophet and basic libraries import karo
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Step 2: Dataset load karo
df = pd.read_csv("monthly_sales.csv")
print(df.head())

# Step 3: Prophet model initialize karo
model = Prophet()

# Step 4: Model ko train karo
model.fit(df)
"""
Future mein kitne months ka forecast chahiye (periods=12  12 months)

freq='MS'  Month Start frequency (jaise 1 Jan, 1 Feb)
"""
# Step 5: Future dates generate karo (next 12 months)
future = model.make_future_dataframe(periods=12, freq='MS')  # MS = Month Start
print(future.tail())
""" Future dataframe ke liye forecast nikal """
# Step 6: Forecast generate karo
forecast = model.predict(future)

# Step 7: Forecast plot karo
model.plot(forecast)
plt.title("Monthly Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# Step 8: Forecast components plot karo (trend, seasonality)
model.plot_components(forecast)
plt.show()
```
